```python
# Application of Scalling method to two independent features
# Feature 1 = 'tenure'
# Feature 2 = 'MonthlyCharges'

from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df['tenure'].max()
```

```python
### Scaling Method 1 
scaler = MinMaxScaler()
X = df[['tenure']]
X_scaled = scaler.fit_transform(X)
print(X_scaled)

# Replace 'tenure' column with Scaled column
df['tenure'] = X_scaled
```

```python
# Scaling Method 2

# Method of MinMax Scaling
y1 = df['MonthlyCharges']
print(type(y1))
print(y1.max())
print(y1.min())
y1
```

```python
# Min-Max Scaling(a=5, b=30)
a = 5
b = 30
y1_new = a + ( (y1 - min(y1))/(max(y1)-min(y1))*(b-a))
print(y1_new.max())
print(y1_new.min())
y1_new
```

```python
# Method 3
# Standardization(Z - SCORE Normalization)
y1_new = (y1-np.mean(y1))/np.std(y1)
print(y1_new.max())
print(y1_new.min())
print(np.mean(y1_new))
print(np.std(y1_new))
```
